## Will work in the flux basis with the Hamiltonian:
## $$ H = \frac12 n_q^2 - \frac{1}{\frac12 + \alpha(t)} \cos(\phi) +  \alpha(t) \cos(2\phi -0.1\pi)$$
## With $\alpha \in [0;1]$

## This looks like:

```python
import torch as t
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 101
phi = t.linspace(-np.pi,np.pi,N)
dphi = np.mean(phi[1:] - phi[:-1])

def V(phi,alpha=1):
    return -1/(0.5+alpha)*t.cos(phi) + alpha*t.cos(2*phi - 0.1*np.pi)

fig, ax = plt.subplots()
for alpha in np.arange(1,0,-0.1):
    ax.plot(phi,V(phi,alpha),c=(0.1,alpha,0.5))
```
